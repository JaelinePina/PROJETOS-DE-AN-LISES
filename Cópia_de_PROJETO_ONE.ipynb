## **Introdução**

O objetivo deste relatório é realizar uma análise detalhada das quatro lojas, com base em dados como faturamento total, média de avaliações dos clientes, volumes de vendas dos produtos e frete médio, para identificar qual loja seria a mais indicada para ser vendida. A análise visa comparar os pontos fortes e fracos de cada loja, com o intuito de recomendar a loja que apresenta o desempenho mais baixo e, consequentemente, a mais propensa à venda.

````python
import pandas as pd
import numpy as np

# Carregar os dados
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Função de limpeza de dados para cada loja
def limpar_dados(loja):
    # Substituir NaN na coluna 'Preço' com a média
    loja['Preço'] = loja['Preço'].fillna(loja['Preço'].mean())
    # Substituir NaN na coluna 'Frete' com a média
    loja['Frete'] = loja['Frete'].fillna(loja['Frete'].mean())

    # Garantir que as colunas 'Preço' e 'Frete' são do tipo float64
    loja['Preço'] = loja['Preço'].astype(np.float64)
    loja['Frete'] = loja['Frete'].astype(np.float64)

    # Remover duplicatas com base nas colunas 'Preço' e 'Frete'
    loja = loja.drop_duplicates(subset=['Preço', 'Frete'])

    return loja

# Limpar os dados para cada loja
loja = limpar_dados(loja)
loja2 = limpar_dados(loja2)
loja3 = limpar_dados(loja3)
loja4 = limpar_dados(loja4)

# Exibir todas as linhas de cada loja para verificar os dados limpos
print("Loja 1:")
print(loja)

print("\nLoja 2:")
print(loja2)

print("\nLoja 3:")
print(loja3)

print("\nLoja 4:")
print(loja4)
````

##**1. Cálculo do Faturamento Total das Lojas**
Primeiramente, foi calculado o faturamento total de cada loja, multiplicando o preço de cada produto pela quantidade vendida. Aqui estão os resultados para cada loja:

Loja 1: O faturamento foi calculado somando os valores de todas as vendas.

Loja 2: Faturamento calculado de forma semelhante.

Loja 3: Faturamento calculado a partir da mesma lógica.

Loja 4: O faturamento foi obtido da mesma forma.

Após calcular o faturamento total de cada loja, podemos observar qual delas gerou maior receita.*

````python
# Cálculo de faturamento: Preço x Quantidade de Parcelas
def calcular_faturamento(loja):
    loja['Faturamento'] = loja['Preço'] * loja['Quantidade de parcelas']
    return loja

# Calcular o faturamento de cada loja
loja = calcular_faturamento(loja)
loja2 = calcular_faturamento(loja2)
loja3 = calcular_faturamento(loja3)
loja4 = calcular_faturamento(loja4)

# Exibir os faturamentos totais de cada loja
faturamento_loja1 = loja['Faturamento'].sum()
faturamento_loja2 = loja2['Faturamento'].sum()
faturamento_loja3 = loja3['Faturamento'].sum()
faturamento_loja4 = loja4['Faturamento'].sum()

print("Faturamento total da Loja 1:", faturamento_loja1)
print("Faturamento total da Loja 2:", faturamento_loja2)
print("Faturamento total da Loja 3:", faturamento_loja3)
print("Faturamento total da Loja 4:", faturamento_loja4)
````

###**4. Produtos Mais e Menos Vendidos**


*    Loja 1:



Produto mais vendido: Micro-ondas com 60 vendas.

Produto menos vendido: Headset com 33 vendas.

*    Loja 2:

Produto mais vendido: Iniciando em programação com 65 vendas.

Produto menos vendido: Jogo de tabuleiro com 32 vendas.

*   Loja 3:

Produto mais vendido: Kit banquetas com 57 vendas.

Produto menos vendido: Blocos de montar com 35 vendas.

*   Loja 4:

Produto mais vendido: Cama box com 62 vendas.

Produto menos vendido: Guitarra com 33 vendas.






````python
import pandas as pd

# Carregar os dados
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Carregar os dados em uma lista
lojas = [pd.read_csv(url) for url in urls]

# Função para limpar dados e garantir que a coluna 'Produto' está limpa
def limpar_dados(loja):
    loja.columns = loja.columns.str.strip()  # Remover espaços extras no nome das colunas
    return loja

# Limpar os dados das lojas
lojas = [limpar_dados(loja) for loja in lojas]

# Função para contar as quantidades de produtos vendidos
def quantidades_vendidos(loja):
    # Contar as ocorrências de cada produto
    vendas = loja['Produto'].value_counts().reset_index()
    vendas.columns = ['Produto', 'Quantidade Vendida']
    return vendas

# Gerar as tabelas de quantidades para cada loja
for i, loja in enumerate(lojas, 1):
    vendas = quantidades_vendidos(loja)
    print(f"\nTabela de Produtos Vendidos - Loja {i}:")
    print(vendas)
````

````python
def quantidades_vendidos(loja):
    # Contar as ocorrências de cada produto
    vendas = loja['Produto'].value_counts().reset_index()
    vendas.columns = ['Produto', 'Quantidade Vendida']
    return vendas

# Identificar os mais e menos vendidos
for i, loja in enumerate(lojas, 1):
    vendas = quantidades_vendidos(loja)

    # Mais vendido (produto com a maior quantidade vendida)
    mais_vendido = vendas.loc[vendas['Quantidade Vendida'].idxmax()]

    # Menos vendido (produto com a menor quantidade vendida)
    menos_vendido = vendas.loc[vendas['Quantidade Vendida'].idxmin()]

    print(f"\nLoja {i}:")
    print(f"Produto mais vendido: {mais_vendido['Produto']} com {mais_vendido['Quantidade Vendida']} vendas.")
    print(f"Produto menos vendido: {menos_vendido['Produto']} com {menos_vendido['Quantidade Vendida']} vendas.")
````

##**5. Frete Médio de Cada Loja**

O frete médio também é um fator importante para a experiência do cliente. Quanto mais barato e eficiente o frete, melhor a avaliação e maior a probabilidade de repetição de compras.

Loja 1: O frete médio foi de [exemplo: R$ 20,00].

Loja 2: O frete médio foi de [exemplo: R$ 15,00].

Loja 3: O frete médio foi de [exemplo: R$ 25,00].

Loja 4: O frete médio foi de [exemplo: R$ 10,00].

Embora o frete mais barato (como no caso da Loja 4) seja um atrativo, a relação entre custo e qualidade do produto também deve ser levada em consideração. A Loja 3, com frete mais caro, pode ter um diferencial em termos de valor agregado.

````python
# Calcular o frete médio de cada loja
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Exibir os resultados
print(f"Frete médio da Loja 1: {frete_medio_loja1:.2f}")
print(f"Frete médio da Loja 2: {frete_medio_loja2:.2f}")
print(f"Frete médio da Loja 3: {frete_medio_loja3:.2f}")
print(f"Frete médio da Loja 4: {frete_medio_loja4:.2f}")
````
